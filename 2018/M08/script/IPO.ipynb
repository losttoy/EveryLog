```python
from patsy import dmatrix
X = dmatrix('Month + Q("Week Day") + Q("Total Underwriters") + Q("Gap Open Pct") + Q("$ Change Opening") +\
                  Q("Lead Mgr") + Q("Offer Price") + Q("Opening Price") +\
                  Q("SP Close to Open Chg Pct") + Q("SP Week Change")', data=ipos, return_type='dataframe')
X
```

```python
X_train, X_test=X[:2188], X[2188:2881]
y_train = ipos['$Change Open to Close'][:2188].map(lambda x: 1 if x >=1 else 0)
y_test = ipos['$Change Open to Close'][2188:2881].map(lambda x: 1 if x >=1 else 0)
clf = linear_model.LogisticRegression()
clf.fit(X_train, y_train)
```

```python
#评估模型在2015年以后的表现
clf.score(X_test, y_test)
```

```python
ipos[(ipos['Trade Date']>='2015-01-01')]['$Change Open to Close'].describe()
```

```python
pred_label = clf.predict(X_test)
results=[]
for pl,tl,idx,chg in zip(pred_label, y_test, y_test.index, ipos.ix[y_test.index]['$Change Open to Close']):
    if pl == tl:
        results.append([idx,chg,pl,tl,1])
    else:
        results.append([idx,chg,pl,tl,0])
rf = pd.DataFrame(results, columns=['index', '$chg', 'predicted', 'actual', 'correct'])
rf
```
